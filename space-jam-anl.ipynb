### Descriçao das colunas
---

**Nome da coluna**|**Descriçao**
:-----------------|-----------------
ID|     Identificador exclusivo de cada jogador no conjunto de dados
points| Total de pontos marcados por um jogador numa época
possessions|    Total de posses de bola de um jogador numa época
team_pace|  Número médio de posses de bola de uma equipa
Unnamed: 4|     Dados desconhecidos
Unnamed: 5|     Dados desconhecidos
GP|     Partidas jogadas por um jogador numa época
MPG|    Média de minutos de jogo de um jogador por partida
TS%|    Percentagem de lançamentos real: a percentagem de lançamentos de um jogador, incluindo lançamentos livres e lançamentos de três pontos
AST|    Taxa de assistências: a percentagem de assistências de um jogador
TO |    Taxa de turnovers: a percentagem de turnovers de um jogador
USG|    Taxa de utilização: o número de posses de bola de um jogador durante 40 minutos
ORR|    Taxa de ressaltos ofensivos
DRR|    Taxa de ressaltos defensivos
REBR|   Taxa de ressaltos: a percentagem de lançamentos falhados com ressalto para um jogador
PER|    Taxa de eficiência dos jogadores, a medida de produtividade por minuto de um jogador em campo

### grafico boxplot 

```python
# importando bibliotecas graficas
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# criando uma list com todos nomes dads colunas menos ID
cols = list(player_df.iloc[:, 1:])
cols
```

```python
# Cria uma matrix 3x5 de subplots
fig, axes = plt.subplots(3,5, figsize=(18, 11))

# Cria preenchimento em torno dos subplots para tornar os rótulos dos eixos
fig.tight_layout(pad=2.0)

# Loop pelas colunas do DataFrame e crie um gráfico de caixa para cada uma.
for i in range(len(cols)):
    sns.boxplot(ax=axes[i//5, i%5], y=player_df[cols[i]])
```
